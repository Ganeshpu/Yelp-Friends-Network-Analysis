```python
# As an user might have more than one friend in the following steps we will unstack the list of friends of each user
# into separate records of a user and his friend
```

```python
new1 = pd.read_csv('user_newidea.csv')
```

```python
new1['friends']=new1["friends"].apply(lambda x: str(x).split(','))
```

```python
new1_fr = pd.concat([new1['user_id'],new1['friends']],axis=1)
```

```python
new1_fr=new1_fr.set_index(['user_id'])['friends'].apply(pd.Series).stack()
```

```python
new1_fr = new1_fr.to_frame().reset_index()
```

```python
new1_fr.pop('level_1')
new1_fr.head()
```

```python
new1_fr.columns = ['user_id','friend_id']
new1_fr.head()
```

```python
new1_fr['friend_id']= new1_fr['friend_id'].str.strip()
```

```python
finallist= new1_fr.user_id.unique().tolist()
filtered = new1_fr.loc[(new1_fr['friend_id'].isin(finallist))]
```

```python
filtered.to_csv('filtered_newidea.csv')
```

```python
#Once we have the file of user and his/her friend we generate the network metrices in 
#Get Network Metrices Using igraph file and import the file below with network metrices and users
```

```python
newnet = pd.read_csv('filtered_newidea_network_metrices.csv')
```

```python
newnet['user_id'] = newnet['vertex']
```

```python
newnet.head()
```

```python
now = pd.merge(newnet,user,how = 'inner', on = 'user_id')
```

```python
now.head()
```

```python
now.columns
```

```python
now_std = now.drop(columns=['user_id','vertex','average_stars','elite','name','friends','yelping_since','Unnamed: 0','funny', 'name', 'review_count', 'useful'],axis=1)
```

```python
now_std
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# In the following steps we are standardizing the network metrices abd user metrices values
std = StandardScaler()
```

```python
std.fit(now_std)
```

```python
std_t = std.transform(now_std)
```

```python
std_t = pd.DataFrame(std_t,columns = now_std.columns)
```

```python
std_t.head()
```

```python
std_t.corr()
```

```python
std_new = now_std.drop(columns=['degree','betweeness','eigenvector'],axis=1)
```

```python
# In below steps we apply PCA on 16 independent variables and get the top 3 components
```

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(n_components=3)
```

```python
pca1.fit(std_new)
```

```python
now_pca1  = pca1.transform(std_new)
```

```python
pca1.components_
```

```python
df_comp1 = pd.DataFrame(pca1.components_,columns=std_new.columns)
```

```python
df_comp1
```

```python
# Here we can see that 95% of the variance in the data is explained by these 3 components.
#Going forward these 3 components will be our independent variables
pca1.explained_variance_ratio_
```

```python
pca_ro = pd.DataFrame(data=now_pca1[0:,0:],columns=['Component_1','Component_2','Component_3'])
```

```python
pca_ro.head()
```

```python
std_s = pd.concat([pca_ro,std_t],axis=1)
```

```python
# Combined user metrices data of 16 independent variables along with the three PCA components.
std_s.head()
```

```python
std_s
```

```python
# As we have 3 dependent variable which we need to predict. If we try to predict each one separately then accuracy
# is not good and as these values are not needed individually we apply PCA on target variable and try to predict it.
```

```python
std_target = pd.concat([std_t['degree'],std_t['betweeness'],std_t['eigenvector']],axis=1)
```

```python
std_target.head()
```

```python
# Getting one componenet by applying PCA on the target variable
pca_target = PCA(1)
```

```python
pca_target.fit(std_target)
```

```python
target_trans = pca_target.transform(std_target)
```

```python
df_target=pd.DataFrame(target_trans,columns = ['Target_PCA'])
```

```python
df_target.head()
```

```python
# Combining all the resultes together with all 16 independent variables their 3 PCA components and 3 dependent
#variables along the one PCA component of the target variable
Final = pd.concat([std_s,df_target],axis=1)
```

```python
Final.to_csv('Final.csv')
```
